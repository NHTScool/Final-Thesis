```python
from tensorflow.keras.preprocessing import image
from tensorflow.keras.models import load_model
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

def image_prediction(new_image_path):
    # Tải và xử lý ảnh
    test_image = image.load_img(new_image_path, target_size=(224, 224))
    test_image = image.img_to_array(test_image)
    test_image = np.expand_dims(test_image, axis=0)
    test_image = test_image / 255.0  # Chuẩn hóa ảnh

    # Load mô hình
    model_loaded = tf.keras.models.load_model("D:\\NHT\\Final Thesis\\code\\codegithub\\my_pneumonia_detection_model.h5")

    # Dự đoán
    prediction = model_loaded.predict(test_image)

    # Hiển thị ảnh
    test_image_for_plotting = image.load_img(new_image_path, target_size=(224, 224))
    plt.imshow(test_image_for_plotting)
    plt.axis('off')  # Ẩn trục để hiển thị đẹp hơn
    plt.show()

    # Xử lý kết quả
    if prediction[0] > 0.5:  # Nếu dự đoán lớn hơn 0.5, kết luận PNEUMONIA
        statistic = prediction[0][0] * 100  # Lấy giá trị xác suất
        print("This image is %.3f percent %s" % (statistic, "P N E U M O N I A"))
    else:  # Nếu không, kết luận NORMAL
        statistic = (1.0 - prediction[0][0]) * 100
        print("This image is %.3f percent %s" % (statistic, "N O R M A L"))

# Gọi hàm để kiểm tra
image_prediction("D:\\NHT\\Final Thesis\\VINAI_Chest_Xray\\test\\test\\0c84dad44979c3a7777a8f4f4d9f5f7a.jpg")
```

```python
from tensorflow.keras.preprocessing import image
from tensorflow.keras.models import load_model
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import os

def folder_image_prediction(folder_path, model_path):
    # Load mô hình
    model_loaded = tf.keras.models.load_model(model_path)
    
    # Lấy danh sách ảnh trong thư mục
    image_files = [f for f in os.listdir(folder_path) if f.endswith(('.jpg', '.png', '.jpeg'))]
    
    if not image_files:
        print("Thư mục không chứa ảnh hợp lệ.")
        return

    for image_file in image_files:
        # Đường dẫn ảnh
        image_path = os.path.join(folder_path, image_file)
        
        # Tải và xử lý ảnh
        test_image = image.load_img(image_path, target_size=(224, 224))
        test_image_array = image.img_to_array(test_image)
        test_image_array = np.expand_dims(test_image_array, axis=0)
        test_image_array = test_image_array / 255.0  # Chuẩn hóa ảnh

        # Dự đoán
        prediction = model_loaded.predict(test_image_array)

        # Hiển thị ảnh
        plt.figure(figsize=(5, 5))
        plt.imshow(test_image)
        plt.axis('off')  # Ẩn trục để hiển thị đẹp hơn
        plt.title(f"Ảnh: {image_file}")
        plt.show()

        # Xử lý kết quả
        if prediction[0][0] > 0.5:  # Nếu dự đoán lớn hơn 0.5, kết luận PNEUMONIA
            statistic = prediction[0][0] * 100  # Lấy giá trị xác suất
            print(f"Ảnh {image_file} là: P N E U M O N I A ({statistic:.3f}%)")
        else:  # Nếu không, kết luận NORMAL
            statistic = (1.0 - prediction[0][0]) * 100
            print(f"Ảnh {image_file} là: N O R M A L ({statistic:.3f}%)")

# Gọi hàm để kiểm tra
folder_path = "D:\\NHT\\Final Thesis\\code\\codegithub\\testanh3"
model_path = "D:\\NHT\\Final Thesis\\code\\codegithub\\my_pneumonia_detection_model.h5"
folder_image_prediction(folder_path, model_path)
```
